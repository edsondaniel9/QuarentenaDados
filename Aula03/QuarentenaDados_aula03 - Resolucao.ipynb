Embora plotamos um monte de informação, não necessariamente reduzimos a correlação em um número para simplificar a análise. Vamos fazer isso com a ajuda do `.corr()` do [pandas](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html). 

```python
imdb_usa[["gross", "budget", "lucro", "title_year"]].corr()
```

Com o pandas é simples de se calcular a correlação, mas precisamos saber interpretar os resultados. Vamos fazer isso?

A correlação é uma métrica que vai de 1 a -1. Quando a correlação é 1, dizemos que é totalmente correlacionada (relação linear perfeita e positiva), ou seja se uma variável aumenta em 10 a outra também irá aumentar em 10. Quando o valor da correlação é -1, também temos variáveis totalmente correlacionda, só que de maneira negativa (relação linear perfeita negativa), neste caso, se uma variável aumenta em 10 a outra reduz em 10. Agora quando a correlação é 0 temos a inexistência de correlação, ou seja, uma variável não tem influêcia sobre a outra. 

Agora sim, entendido sobre a correlação vamos analisar as nossas. Veja que lucro e gross tem uma correlação alta, o que indica que quanto maior o orçamento maior o lucro (mas repare que a correlação não é perfeita), já o title_yers e lucro tem correlação negativa, mas muito perto de zero (ou seja quase não tem correlação). Em aula falamos que talvez, filmes mais recentes podem ter prejuizo pois ainda não tiveram tempo de recuperar o investimento. Analise essas informações e nos conte quais foram suas conclusões.

```python
print("Intervalo de anos dos filmes de {} até {}".format(imdb_usa.title_year.min(), imdb_usa.title_year.max()))
```

```python
def plot_antes_e_depois_ano(ano_corte, lim_inferior):
    budget_lucro_r = \
            imdb_usa.query("(budget >0 or gross > 0) and title_year >= @ano_corte and title_year >= @lim_inferior")\
                                    [["budget", "lucro", "title_year"]].dropna()
    budget_lucro_o = \
            imdb_usa.query("(budget >0 or gross > 0) and title_year < @ano_corte and title_year >= @lim_inferior")\
                                    [["budget", "lucro", "title_year"]].dropna()

    f, axs = plt.subplots(1, 2, sharey=True, figsize=(22, 5))

    plot_recent = sns.scatterplot(x="title_year", y="lucro", data=budget_lucro_r, ax=axs[0])
    plot_recent.set_title("Distribuição de {}".format(ano_corte))

    plot_old = sns.scatterplot(x="title_year", y="lucro", data=budget_lucro_o, ax=axs[1])
    plot_old.set_title("Distribuição antes de {}".format(ano_corte))

    plt.show()

plot_antes_e_depois_ano(2016, 1900)
```

```python
def print_lucro_usa(ano):
    df_local = df_agrup_filmes_por_ano.query("title_year >= @ano")

    # Plotar o grafico de lucro por ano (medio por hora), avaliaremos apenas anos com valores
    df_agrup_filmes_por_ano_media = df_local.groupby('title_year').mean()
    df_agrup_filmes_por_ano_mediana = df_local.groupby('title_year').median()

    f, axs = plt.subplots(1, 2, sharey=True, figsize=(22, 5))

    # Plot por media
    plot_media = df_agrup_filmes_por_ano_media.lucro.plot(ax=axs[0])
    plot_media.set_title("Distribuição das Médias")

    plot_mediana = df_agrup_filmes_por_ano_mediana.lucro.plot(ax=axs[1])
    plot_mediana.set_title("Distribuição das Medianas")

    plt.show()
    # display(df_agrup_filmes_por_ano_media.tail())
    # display(df_agrup_filmes_por_ano_mediana.tail())
    
print_lucro_usa(1969)
```
